```python
import os
import sys
import shutil

from pypost.experiments.Experiment import Experiment
from pypost.tutorials.experiments.DummyTrial import DummyTrial
from pypost.plotter.PlotterEvaluations import PlotterErrorBars
from pypost.plotter.PlotterEvaluations import PlotterTrials
```

Create Experiment in the directory with given name. Execute DummyTrial class

Afterwards, we create two evaluation collections and evaluate 3 trials per parameter set. If the experiment is not there yet, it will be created and the trials are evaluated. Otherwise, we will load the experiment data. 

```python
experiment = Experiment('/tmp', 'testCategory', DummyTrial)

experiment.defaultSettings.temperature = 2
experiment.create()

evaluation1 = experiment.addEvaluationCollection(['temperature'], [0.1, 1, 2, 3], numTrials = 3)
evaluation2 = experiment.addEvaluationCollection(['numIterations'], [100, 200, 300], numTrials = 3)

experiment.startLocal()
```

```python
We can get the data from an evaluation collection. The data is a standard pypost core data object 
containing the learning curves and the parameter values of the trials
```

```python
%matplotlib notebook
data = experiment.getEvaluationData(evaluation1)
```

The data can be accesed in the standard way

```python
print(data.getNumElementsForDepth(0))
data[0,1].avgReturns, data[...].temperature
```

```python
data.printDataAliases()
```

We can now plot our evaluation collections

```python
plotter = PlotterErrorBars(data.dataManager, 'avgReturns', 'temperature', useMedian=True, logSpaceY=True)
data[[1,2], ..., ...] >> plotter
```

This also works in log-space and if we have a different number of iterations

```python
data = experiment.getEvaluationData(evaluation2)

plotter = PlotterErrorBars(data.dataManager, 'avgReturns', 'numIterations', useMedian=True, logSpaceY=True, useEpisodesXLabel=True)
data[...] >> plotter
```

```python
We can also plot the individual trials
```

```python
plotter = PlotterTrials(data.dataManager, 'avgReturns', 'numIterations', logSpaceY=True, useEpisodesXLabel=True, legendPerTrial=True)
data[...] >> plotter
```
